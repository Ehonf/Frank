```python
import matplotlib.pyplot as plt

# Données
pays = ['Sénégal', 'Mali', 'Guinée']
population = [17, 22, 13]  # en millions

# Question 1 : Que fait cette ligne ?
fig, ax = plt.subplots()
# Réponse :  Elle crée une nouvelle figure (la fenêtre) et un système d'axes (le graphique) pour pouvoir dessiner.

# Question 2 : Que va afficher cette ligne ?
ax.bar(pays, population, color=['green', 'yellow', 'red'])
# Réponse :  Elle affiche un diagramme en barres avec les pays en abscisse, leurs populations en ordonnée, et des barres aux couleurs du drapeau (vert, jaune, rouge).

# Question 3 : Quel est le rôle de ax.set_ylabel ?
ax.set_ylabel('Population (millions)')
# Réponse : Elle permet d'ajouter une étiquette (un nom) sur l'axe vertical (ordonnées) pour expliquer ce que les chiffres représentent.

ax.set_title('Population de trois pays ouest-africains')
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Produit A', 'Produit B', 'Produit C', 'Produit D', 'Produit E', 
              'Produit F', 'Produit G', 'Produit H', 'Produit I', 'Produit J']
ventes = [150, 220, 180, 300, 250, 50, 45, 30, 25, 20]

# Création de la figure
fig, ax = plt.subplots(figsize=(10, 6))

# Utilisation de barres horizontales pour la lisibilité
ax.barh(categories, ventes, color='skyblue')

# Inverser l'axe pour avoir le Produit A en haut
ax.invert_yaxis()

# Ajout des labels et du titre
ax.set_xlabel('Nombre de ventes')
ax.set_ylabel('Produits')
ax.set_title('Classement des ventes par produit')

# Optionnel : ajout d'une grille pour faciliter la lecture des valeurs
ax.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

mpg = sns.load_dataset('mpg')

plt.figure(figsize=(10, 6))
sns.scatterplot(data=mpg, x='weight', y='mpg', hue='origin')
plt.title('Relation entre weight et mpg')
plt.show()
```

```python
result = mpg.groupby('origin')['mpg'].mean()
print(result)
```

```python
res = mpg.groupby('origin')['mpg'].median()
print(res)
```
