```python
"Question 2.2"
correction = {'ALICE':'Alice', 'BOB':'Bob', 'alice':'Alice'}
ventes_sales['vendeur'] = ventes_sales['vendeur'].replace(correction)
print(ventes_sales['vendeur'])
```

```python
ventes_sales.dropna(subset=['date']) 
```

```python
print(ventes_sales.duplicated().sum())
```

```python
ventes_sales.drop_duplicates(inplace=True)
```

```python
ventes_sales['prix_unit'] = ventes_sales['prix_unit'].fillna(ventes_sales.groupby('produit')['prix_unit'].transform('median'))
```

```python
ventes_sales['quantite'] = ventes_sales['quantite'].clip(lower=0)
medians = ventes_sales.groupby('produit')['quantite'].median().astype(int)
ventes_sales.loc[ventes_sales['quantite'] > 200, 'quantite'] = ventes_sales.loc[ventes_sales['quantite'] > 200, 'produit'].map(medians)
print(ventes_sales['quantite'])
```

```python
total = ventes_sales['quantite'] * ventes_sales['prix_unit']
print(total)
```

```python
ventes = pd.DataFrame({
'vendeur': ['Alice','Bob','Alice','Bob','Alice','Charlie','Charlie'],
'produit': ['Riz','Huile','Riz','Tomate','Riz','Sucre','Sucre'],
'quantite':[10, 5, 8, 20, 10, 15, 12],
'prix_unit':[500, 1200, 500, 250, 500, 600, 600],
'total': [5000, 6000, 4000, 5000, 5000, 9000, 7200],
})
print(ventes)
```

```python

analyse_vendeur = ventes.groupby('vendeur')[['quantite', 'total']].sum()
print("a) Analyse par vendeur :\n", analyse_vendeur)
```

```python
pivot = ventes.pivot_table(values='total', index='vendeur', columns='produit', aggfunc='sum', fill_value=0)
print(pivot)
```

```python
ventes['ca_vendeur'] = ventes.groupby('vendeur')['total'].transform('sum')
print("\nc) DataFrame avec CA vendeur :\n", ventes)
```

```python
print(ventes['ca_vendeur'])
```

```python
ventes['part_ca'] = (ventes['total'] / ventes['ca_vendeur']) * 100
print("\nd) DataFrame avec part du CA (%) :\n", ventes)
```

```python
vendeurs_info = pd.DataFrame({
'vendeur': ['Alice', 'Bob', 'Charlie', 'Diana'],
'region': ['Littoral', 'Centre', 'Centre', 'Ouest'],
'anciennete_an': [3, 7, 1, 5],
})
objectifs = pd.DataFrame({
'vendeur': ['Alice', 'Bob', 'Charlie'],
'objectif_ca': [20000, 25000, 15000],
})
```

```python
print(vendeurs_info)
```

```python
# Fusion des ventes avec les infos vendeurs
df_fusion = pd.merge(ventes, vendeurs_info, on='vendeur', how='left')
```

```python
# On utilise la table complète des vendeurs pour ne perdre personne
df_final = pd.merge(df_fusion, objectifs, on='vendeur', how='left')
```

```python
# 1. Calcul du CA total par vendeur sur le DataFrame fusionné
ca_par_vendeur = df_final.groupby('vendeur')[['total', 'objectif_ca']].first().reset_index()

# 2. Calcul du taux
ca_par_vendeur['taux_realisation'] = (ca_par_vendeur['total'] / ca_par_vendeur['objectif_ca']) * 100

# 3. Gestion du cas Diana (remplacer NaN par 0 ou un message)
ca_par_vendeur['taux_realisation'] = ca_par_vendeur['taux_realisation'].fillna(0)
print(ca_par_vendeur[['vendeur', 'total', 'objectif_ca', 'taux_realisation']])
```
